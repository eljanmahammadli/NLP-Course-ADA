```python
def word_tokenize_special(sentence):
  strings = [word for word in word_tokenize(sentence) if word not in special_characters]
  azerbaijani_alphabet = "AaBbCcÇçDdEeƏəFfGgĞğHhXxIıİiJjKkQqLlMmNnOoÖöPpRrSsŞşTtUuÜüVvYyZz"
  pattern = f"^[{azerbaijani_alphabet}]+$"
  filtered_strings = [s for s in strings if re.match(pattern, s)]
  filtered_strings = [s.lower() for s in filtered_strings]
  return filtered_strings
```

```python

b = {}
sentences = sent_tokenize(combined_text)
for sent in sentences:
  wrds = ['<BOS>'] + word_tokenize_special(sent) + ['<EOS>']
  for w1, w2 in zip(wrds, wrds[1:]):
    bigram = (w1, w2)
    b[bigram] = b.get(bigram, 0) + 1
```

```python
words = set()
for sent in sentences:
    wrds = word_tokenize_special(sent)
    for w in wrds:
        words.add(w)
special_toks = ['<BOS>', '<EOS>']
words = list(words)
vocab_size = len(words) + len(special_toks)
N = torch.zeros((vocab_size, vocab_size), dtype=torch.int32)
stoi = {s:i+2 for i,s in enumerate(words)}
stoi['<EOS>'] = 0
stoi['<BOS>'] = 1
itos = {i:s for s,i in stoi.items()}
```

```python
# for sent in sentences[3000: 3005]:
for sent in sentences:
  wrds = ['<BOS>'] + word_tokenize_special(sent) + ['<EOS>']
  for w1, w2 in zip(wrds, wrds[1:]):
    ix1 = stoi[w1]
    ix2 = stoi[w2]
    N[ix1, ix2] += 1
    # print(w1, w2)
```

```python
P = (N+10).float() # laplace smoothing
P /= P.sum(1, keepdims=True)
```

```python
g = torch.Generator().manual_seed(10110609)

for i in range(1):
  
  out = []
  ix = 1
  while True:
    p = P[ix]
    ix = torch.multinomial(p, num_samples=1, replacement=True, generator=g).item()
    out.append(itos[ix])
    if ix == 0:
      break
  print(' '.join(out))
```

```python
log_likelihood = 0.0
n = 0

for sent in sentences:
  wrds = ['<BOS>'] + word_tokenize_special(sent) + ['<EOS>']
  for w1, w2 in zip(wrds, wrds[1:]):
    ix1 = stoi[w1]
    ix2 = stoi[w2]
    prob = P[ix1, ix2]
    logprob = torch.log(prob)
    log_likelihood += logprob
    n += 1
    #print(f'{ch1}{ch2}: {prob:.4f} {logprob:.4f}')

print(f'{log_likelihood=}')
nll = -log_likelihood
print(f'{nll=}')
print(f'{nll/n}')
perp = torch.exp(nll/n)
```
